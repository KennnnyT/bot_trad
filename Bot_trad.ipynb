```python
# Importer les bibliothèques nécessaires
import ccxt
import time

# Définir les variables d'environnement
exchange = ccxt.binance()
symbol = 'BTC/USDT'
amount = 1

# Fonction pour acheter du BTC à un prix donné
def buy_btc(price):
  order = exchange.create_order(symbol, 'limit', 'buy', amount, price)
  print(f'Achat de {amount} BTC à un prix de {price} USDT')

# Fonction pour vendre du BTC à un prix donné
def sell_btc(price):
  order = exchange.create_order(symbol, 'limit', 'sell', amount, price)
  print(f'Vente de {amount} BTC à un prix de {price} USDT')

# Boucle infinie pour surveiller le prix du BTC
while True:
  # Récupérer le prix actuel du BTC
  ticker = exchange.fetch_ticker(symbol)
  price = ticker['last']
  print(f'Prix actuel du BTC : {price} USDT')

  # Si le prix du BTC est supérieur à 10 000 USDT, vendre
  if price > 10000:
    sell_btc(price)

  # Si le prix du BTC est inférieur à 9 000 USDT, acheter
  elif price < 9000:
    buy_btc(price)

  # Attendre 1 minute avant de vérifier à nouveau le prix
  time.sleep(60)
```
